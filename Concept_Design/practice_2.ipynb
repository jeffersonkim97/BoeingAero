Design variables for the system are altitude H, framerate fr, sweep area A.

```python
# Import initialization:
import casadi as ca
from casadi import *
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from casadi.tools.graph import dotgraph
from IPython.display import Image

def draw_graph(expr):
    return Image(dotgraph(expr).create_png())
```

```python
# Unit Conversion
lbs2kg = 0.453592 # lbs to kg (Weight)
slugft2kgm = 1.35581795 # slug ft^2 to kg m^2 (Intertia)
mi2km = 1.60934 # Mi to Km (Distance)
```

```python
# Camera Data
data = {'Hero5_Black': {'f': {'wide': 14, 'medium': 21, 'narrow': 28},
                        'sensor_size_h': 6.17, 'sensor_size_v': 4.55, 'Weight': 118,
                        'video_resolution': {'4K': {'wide': 30},
                                             '2.7K': {'wide': 60, 'medium': 48, 'narrow': 24},
                                             '1440p': {'wide': 60},
                                             '1080p': {'wide': 60, 'medium': 48, 'narrow': 30}}}
       } # Units: f: mm, sensor size: mm, Weight: g
```

```python
def sensor_spec(model_name):
    '''
    This function takes name of camera model and output its specficiation
    
    Input: model_name (str)
    Output: specification (dic)
    '''
    specification = data[model_name]
    return specification

def sensor_calculation(specification, cam_type, H, D, VT):    
    # Sensor Model:
    f = float(specification['f'][cam_type])*1e-3 # Focal Lnegth (m)
    h = float(specification['sensor_size_h'])*1e-3 # Lense Height (m)
    P = 0.5 # Percentage of required forward overlap    

    AFOV = 2*np.arctan(h/(2*f))
    S = 2*H*np.tan(AFOV/2)
    frame_rate = VT*(S*(1 - P))
    t_exp = 1 # Flight Time (sec)
    Width = VT*t_exp
    A = S*Width
    
    return A
```

```python
H = np.linspace(0, 120, 1000)
D = 12.5 * mi2km
VT = np.linspace(0.000000001, 50, 10)

sensor = sensor_spec('Hero5_Black')
cam_type = 'wide'

A0 = sensor_calculation(sensor, cam_type, H, D, 0.01)
A1 = sensor_calculation(sensor, cam_type, H, D, 5)
A2 = sensor_calculation(sensor, cam_type, H, D, 10)
A3 = sensor_calculation(sensor, cam_type, H, D, 15)

fig, ax = plt.subplots()
ax.plot(H, A0, label='VT=0.01')
ax.plot(H, A1, label='VT=5')
ax.plot(H, A2, label='VT=10')
ax.plot(H, A3, label='VT=15')
leg = ax.legend();
plt.xlabel('Height (m)')
plt.ylabel('Sweep Area (m^2)')
plt.grid()
plt.show()

#for i in VT:
#    A = sensor_calculation(sensor, cam_type, H, D, i)
#    plt.plot(H, A, label=i)
#plt.xlabel('Height (m)')
#plt.ylabel('Sweep Area (m^2)')
#plt.grid()
#plt.show()
```

```python
# At given height range, determine focal length and sensor height
# Design Vector: altitude H, speed VT
# Output: focal length f, sensor height h, frame rate fr, overlap percentage P

def constraint(s, param):
    H = param[0]
    D = param[1] * mi2km
    f = param[2]
    h = param[3]
    
    VT = s[0]
    fr = s[1]
    P = s[2]
    
    u = ca.vertcat(H, VT)
    x = ca.vertcat(fr, P)
    p = ca.vertcat(f, h, D)
    
    return x, u, p

def sensor_requirements(x, u, p):
    H = u[0]
    VT = u[1]
    
    
    fr = x[0]
    P = x[1]
    
    f = p[0]
    h = p[1]
    D = p[2]
    
    AFOV = 2*H*ca.atan(h/(2*f))
    S = 2*H*ca.tan(AFOV/2)
    fr = VT*(S*(1 - P))
    t_exp = D / VT # Flight Time (sec)
    Width = VT*t_exp
    A = S*Width
    
    output = ca.vertcat(fr, A)
    
    return output
```

```python
H = ca.SX.sym('H')
D = ca.SX.sym('D')
f = ca.SX.sym('f')
h = ca.SX.sym('h')
param = ca.vertcat(H, D, f, h)


VT = ca.SX.sym('VT')
fr = ca.SX.sym('fr')
P = ca.SX.sym('P')

s = ca.vertcat(VT, fr, P)
```

```python
x, u, p = constraint(s, param)
output = sensor_requirements(x, u, p)
```

```python
draw_graph(output)
```

```python
def trim_cost(x, u, output):
    H = u[0]
    VT = u[1]
    
    fr = x[0]
    P = x[1]
    
    frame_rate = output[0]
    A = output[1]
    
    return (1/A)**2 + 1e4*(1/P)**2 + (frame_rate)**2
```

```python
trim_cost_val = trim_cost(x, u, output)
```

```python
draw_graph(trim_cost_val)
```

```python
def objective(s, param):
    x, u, p = constraint(s, param)
    output = sensor_requirements(x, u, p)
    c = trim_cost(x, u, output)
    return c
```

```python
c = objective(s, param)

draw_graph(c)
```

```python
def sensor_requirement_opt(x0, param):
    x_design = ca.SX.sym('x_design', 3)
    trim_cost = objective(x_design, param)
    nlp = {'x': x_design, 'f': trim_cost}
    S = nlpsol('S', 'ipopt', nlp)
    result = S(x0=x0, lbg=0, ubg=0,
               lbx=[0.1, 1, 0.01],
               ubx=[20, 100, 0.9])
    x_opt = result['x']
    return x_opt
```
